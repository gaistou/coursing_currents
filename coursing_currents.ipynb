```python
import random

def simulate_event(base_cs_chance):
    bonus = 0
    attempts = 0

    while True:
        # print(f"attempt {attempts + 1}")
        attempts += 1
        cs_chance = base_cs_chance + (attempts - 1) * 0.03

        if random.random() < cs_chance:
            return bonus

        # print(bonus)
        bonus += 0.03

def main(num_trials, base_cs_chance):
    total_bonus = 0
    for _ in range(num_trials):
        total_bonus += simulate_event(base_cs_chance)

    average_bonus = total_bonus / num_trials
    return average_bonus

num_trials = 100000

for i in range(50):
    average_bonus = main(num_trials, i/100)
    print(f"Average Bonus over {num_trials} trials for {i}% base_cs_chance: {(average_bonus * 100):.4f}")
```
